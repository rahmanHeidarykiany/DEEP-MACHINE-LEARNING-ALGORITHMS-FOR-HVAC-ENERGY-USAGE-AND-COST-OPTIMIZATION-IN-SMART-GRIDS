```python
# evaluate one or more weekly forecasts against expected values
def evaluate_forecasts(actual, predicted):
    scores = list()
    # calculate an RMSE score for each day;
    # actual.shape[1] should be 24 in our case;
    for i in range(actual.shape[1]):
        # calculate mse
        mse = mean_squared_error(actual[:, i], predicted[:, i])
        # calculate rmse
        rmse = sqrt(mse)
        # store
        scores.append(rmse)
    # calculate overall RMSE
    # actual.shape[0] should be 26 in our case - as the number of days in
    # the test set;
    s = 0
    for row in range(actual.shape[0]):
        for col in range(actual.shape[1]):
            s += (actual[row, col] - predicted[row, col])**2
    score = sqrt(s / (actual.shape[0] * actual.shape[1]))
    return score, scores

```

```python

# evaluate one or more forecasts against expected values
# as relative errors expressed as percentages
# 100 * (predicted - actual)/actual
# in other  words, report mean absolute percentage error (MAPE)
# which is commonly used to measure the predictive accuracy of models
def percentage_error(actual, predicted):
    res = np.empty(actual.shape)
    for j in range(actual.shape[0]):
        if actual[j] != 0:
            res[j] = (actual[j] - predicted[j]) / actual[j]
        else:
            res[j] = predicted[j] / np.mean(actual)
    return res
```

```python
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs(percentage_error(np.asarray(y_true), np.asarray(y_pred)))) * 100

def evaluate_forecasts_MAPE(actual, predicted):
    print("#########################################################################")
    print("#################evaluate_forecasts_MAPE##########################################")
    print(" ")
    scores = list()
    # actual.shape[1] should be 24 in our case;
    for i in range(actual.shape[1]):
        # calculate mape for this hour
        mape = mean_absolute_percentage_error(actual[:, i], predicted[:, i])
        print("mape")
        print(mape)
        scores.append(mape)
    # calculate overall average MAPE over all 24 hours
    score = sum(scores)/len(scores)
    return score, scores

# summarize scores

# summarize scores
def summarize_scores(score_MAPE, scores_MAPE, score_RMSE,scores_RMSE):
      print('RMSE　Result')
      s_scores = ', '.join(['%.1f' % s for s in scores_RMSE])
      name='lstm_RMSEs'
      #print('%s: [%.3f] %s' % (name, score, s_scores))
      print('%s: %s' % (name, scores_RMSE))
      print(" ")
      print(" calculate overall average RMSE over all 24 hours = {} ".format(score_RMSE))
      print(" ")
      print('MAPE　Result')
      s_scores = ', '.join(['%.1f' % s for s in scores_MAPE])
      name='lstm_MAPEs'
      #print('%s: [%.3f] %s' % (name, score, s_scores))
      print('%s: %s' % (name, scores_MAPE))
      print(" ")
      print(" calculate overall average MAPE over all 24 hours = {} ".format(score_MAPE))
      print(" ")
```

```python

# convert history into inputs and outputs
def to_supervised(train, n_input, n_out=24): # =24, =7####### n_out=1 means always one y_lable will be there
    # flatten data
    data = train.reshape((train.shape[0]*train.shape[1], train.shape[2]))
    print("in func to_supervised; {} is data.shape (train.shape after reshape) ".format(data.shape))
    print("#########################################################################")
    print("#################to_supervised##########################################")
    print(" ")
    print("data[1]")
    print(data[1])
    print("data[10]")
    print(data[10])
    print("data[11]")
    print(data[11])
    N_days_for_training_data=int(data.shape[0]/24)
    print("Number of days for training data =  {} days".format(N_days_for_training_data))
    X, y = list(), list()
    in_start = 0
    # step over the entire history one time step at a time
    for _ in range(N_days_for_training_data):
        # define the end of the input sequence
        in_end = in_start + n_input
        out_end = in_end + n_out
        # ensure we have enough data for this instance
        if out_end <= len(data):
            # take the first 24 rows, only the 1-st column index 0
            # (which is total Net_total) of the data that has shape
            # (8400,6); that is "total calculated power" each hour,
            # for 24 hours, i.e., 1 days;
            x_input = data[in_start:in_end, 1] ### 0 is "Net total", 1, is "Lights", 2 is "HVAC"
            x_input = x_input.reshape((len(x_input), 1))
            X.append(x_input)
            y.append(data[in_end:out_end, 1]) ### 0 is "Net total", 1, is "Lights", 2 is "HVAC"
        else:
          print(" out_end > len(data) and out_end: {} ".format(out_end))
        # move along one time step
        # NOTE: when prediction was done for 7 days values, incrementing
        # one day at a time here was ok; it was sun-sat, mon-sun, tue-mon,
        # wed-tue, etc.
        # this was done like that to increase amount of train data; it was
        # ok because the daily total consumption does not vary that much;
        # however, when doing 24 h for 24 values prediction, incrementing
        # hourly for the sake of increasing amount of train data makes for the
        # outputs to predict "average values" because training of model is done
        # with data 0-23h, 1-0h, 2-1h, 3-2h etc. instead of
        # 0-23h, 0-23h, 0-23h, 0-23h, etc.
        # overlapping days I think diminishes the "daily pattern" of power
        # and kind-of-uniformizes it;
        in_start += 24 # 24
        print(" ")
        print("**********************************")
        print("*************** finish for loop in to_supervised **************")
        print(" len of list X =  {} ".format(len(X)))
        print(" len of list y =  {} ".format(len(y)))
    return array(X), array(y)
```

```python
import numpy as np

n_input = 9
n_out = 9

def to_supervised_RNN(train, n_input, n_out=n_out, target_column=1):
    data = train.reshape((train.shape[0] * train.shape[1], train.shape[2]))
    N_days_for_training_data = int(data.shape[0] / 9)
    X, Y = list(), list()
    in_start = 0
    for _ in range(N_days_for_training_data):
        in_end = in_start + n_input
        out_end = in_end + n_out
        if out_end <= len(data):
            x_input = data[in_start:in_end, :]
            X.append(x_input[:, [col for col in range(data.shape[1]) if col != target_column]])
            y_input = data[in_end:out_end, target_column].reshape(-1, 1)  # Select the desired column and reshape to (n_out, 1)
            Y.append(y_input)
        in_start += n_input

    return np.array(X), np.array(Y)

# Example usage
# Assuming you have a variable named "train" that contains your data
train_input, train_target = to_supervised_RNN(train, n_input, n_out, target_column=1)
test_input,test_target = to_supervised_RNN(test, n_input, n_out, target_column=1)

# Now, train_input will not contain the "target_column" information.
# train_input shape will be (899, 9, 5)
# train_target shape will be (899, 9, 1), containing only the "target_column" information.
print(train_input.shape, train_target.shape)
print(test_input.shape, test_target.shape)
```

```python
def evaluate_model_2(model, test_input, test_target, sequence_length, input_size):
    model.eval()
    with torch.no_grad():
        prediction_seqs = []
        cvrmse_sum = 0.0
        for i in range(len(test_input)):
            input_seq = test_input[i:i+1]
            target_seq = test_target[i:i+1]

            input_expanded = input_seq.reshape(1, sequence_length, input_size)
            hidden_state = torch.zeros(1, model.hidden_size)
            cell_state = torch.zeros(1, model.hidden_size)  # Fix batch size to 1 for LSTM
            input_t_past = torch.zeros_like(input_expanded[:, 0])
            prediction_seq = []
            squared_errors = []

            for t in range(sequence_length):
                output, _, hidden_state = model(input_expanded[:, t], hidden_state, cell_state, input_t_past)
                prediction = output[0, 0]
                prediction_seq.append(output[0, 0])  # Append the model output directly
                input_t_past = input_expanded[:, t]
                squared_error = (prediction - target_seq[:, t]) ** 2
                squared_errors.append(squared_error)

            prediction_seq = np.array(prediction_seq)
            prediction_seqs.append(prediction_seq)

            squared_errors = np.array(squared_errors)
            rmse = np.sqrt(np.mean(squared_errors.astype(np.float64)))  # Convert to appropriate data type
            target_mean = torch.mean(target_seq).item()  # Convert tensor to float
            cvrmse = rmse / target_mean
            cvrmse_sum += cvrmse

        average_cvrmse = cvrmse_sum / len(test_input)
        #print("Average CVRMSE on Test Data:", average_cvrmse)
        return np.array(prediction_seqs), average_cvrmse
```

```python
import torch
import numpy as np

def evaluate_model_mape(model, test_input, test_target, sequence_length, input_size):
    model.eval()
    with torch.no_grad():
        prediction_seqs = []
        mape_sum = 0.0
        for i in range(len(test_input)):
            input_seq = test_input[i:i+1]
            target_seq = test_target[i:i+1]

            input_expanded = input_seq.reshape(1, sequence_length, input_size)
            hidden_state = torch.zeros(1, model.hidden_size)
            cell_state = torch.zeros(1, model.hidden_size)  # Fix batch size to 1 for LSTM
            input_t_past = torch.zeros_like(input_expanded[:, 0])
            prediction_seq = []
            absolute_percentage_errors = []

            for t in range(sequence_length):
                output, _, hidden_state = model(input_expanded[:, t], hidden_state, cell_state, input_t_past)
                prediction = output[0, 0]
                prediction_seq.append(output[0, 0])  # Append the model output directly
                input_t_past = input_expanded[:, t]
                # Calculate absolute percentage error
                absolute_percentage_error = torch.abs((prediction - target_seq[:, t]) / target_seq[:, t])
                absolute_percentage_errors.append(absolute_percentage_error)

            prediction_seq = np.array(prediction_seq)
            prediction_seqs.append(prediction_seq)

            absolute_percentage_errors = torch.cat(absolute_percentage_errors)
            mape = torch.mean(absolute_percentage_errors).item() * 100  # Convert to percentage
            mape_sum += mape

        average_mape = mape_sum / len(test_input)
        return np.array(prediction_seqs), average_mape
```

```python

class LSTM(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(LSTM, self).__init__()
        self.hidden_size = hidden_size
        ## 1
        self.W_ii_1 = nn.Linear(input_size, hidden_size, bias=False)
        self.W_ii_1.weight.data = torch.abs(self.W_ii_1.weight.data).float()
        self.W_hi_1 = nn.Linear(hidden_size, hidden_size, bias=False)
        self.W_hi_1.weight.data = torch.abs(self.W_hi_1.weight.data).float()
        self.b_i_1 = nn.Parameter(torch.zeros(1, hidden_size))

        self.W_if_1 = nn.Linear(input_size, hidden_size, bias=False)
        self.W_if_1.weight.data = torch.abs(self.W_if_1.weight.data).float()
        self.W_hf_1 = nn.Linear(hidden_size, hidden_size, bias=False)
        self.W_hf_1.weight.data = torch.abs(self.W_hf_1.weight.data).float()
        self.b_f_1 = nn.Parameter(torch.zeros(1, hidden_size))

        self.W_ig_1 = nn.Linear(input_size, hidden_size, bias=False)
        self.W_ig_1.weight.data = torch.abs(self.W_ig_1.weight.data).float()
        self.W_hg_1 = nn.Linear(hidden_size, hidden_size, bias=False)
        self.W_hg_1.weight.data = torch.abs(self.W_hg_1.weight.data).float()
        self.b_g_1 = nn.Parameter(torch.zeros(1, hidden_size))

        self.W_io_1 = nn.Linear(input_size, hidden_size, bias=False)
        self.W_io_1.weight.data = torch.abs(self.W_io_1.weight.data).float()
        self.W_ho_1 = nn.Linear(hidden_size, hidden_size, bias=False)
        self.W_ho_1.weight.data = torch.abs(self.W_ho_1.weight.data).float()
        self.b_o_1 = nn.Parameter(torch.zeros(1, hidden_size))

        self.D1_1 = nn.Linear(input_size, hidden_size, bias=False)
        self.D1_1.weight.data = torch.abs(self.D1_1.weight.data).float()
        self.D2_1 = nn.Linear(input_size, hidden_size, bias=False)
        self.D2_1.weight.data = torch.abs(self.D2_1.weight.data).float()
        self.D3_1 = nn.Linear(input_size, hidden_size, bias=False)
        self.D3_1.weight.data = torch.abs(self.D3_1.weight.data).float()
        self.D4_1 = nn.Linear(input_size, hidden_size, bias=False)
        self.D4_1.weight.data = torch.abs(self.D4_1.weight.data).float()
        ##
        ##
        self.D5_1 = nn.Linear(hidden_size, hidden_size)
        self.D5_1.weight.data = torch.abs(self.D5_1.weight.data).float()
        self.D6_1 = nn.Linear(input_size, hidden_size, bias=False)
        self.D6_1.weight.data = torch.abs(self.D6_1.weight.data).float()
        ##
        ##
        ## _2
        self.W_ii_2 = nn.Linear(input_size, hidden_size, bias=False)
        self.W_ii_2.weight.data = torch.abs(self.W_ii_2.weight.data).float()
        self.W_hi_2 = nn.Linear(hidden_size, hidden_size, bias=False)
        self.W_hi_2.weight.data = torch.abs(self.W_hi_2.weight.data).float()
        self.b_i_2 = nn.Parameter(torch.zeros(1, hidden_size))

        self.W_if_2 = nn.Linear(input_size, hidden_size, bias=False)
        self.W_if_2.weight.data = torch.abs(self.W_if_2.weight.data).float()
        self.W_hf_2 = nn.Linear(hidden_size, hidden_size, bias=False)
        self.W_hf_2.weight.data = torch.abs(self.W_hf_2.weight.data).float()
        self.b_f_2 = nn.Parameter(torch.zeros(1, hidden_size))

        self.W_ig_2 = nn.Linear(input_size, hidden_size, bias=False)
        self.W_ig_2.weight.data = torch.abs(self.W_ig_2.weight.data).float()
        self.W_hg_2 = nn.Linear(hidden_size, hidden_size, bias=False)
        self.W_hg_2.weight.data = torch.abs(self.W_hg_2.weight.data).float()
        self.b_g_2 = nn.Parameter(torch.zeros(1, hidden_size))

        self.W_io_2 = nn.Linear(input_size, hidden_size, bias=False)
        self.W_io_2.weight.data = torch.abs(self.W_io_2.weight.data).float()
        self.W_ho_2 = nn.Linear(hidden_size, hidden_size, bias=False)
        self.W_ho_2.weight.data = torch.abs(self.W_ho_2.weight.data).float()
        self.b_o_2 = nn.Parameter(torch.zeros(1, hidden_size))

        self.D1_2 = nn.Linear(input_size, hidden_size, bias=False)
        self.D1_2.weight.data = torch.abs(self.D1_2.weight.data).float()
        self.D2_2 = nn.Linear(input_size, hidden_size, bias=False)
        self.D2_2.weight.data = torch.abs(self.D2_2.weight.data).float()
        self.D3_2 = nn.Linear(input_size, hidden_size, bias=False)
        self.D3_2.weight.data = torch.abs(self.D3_2.weight.data).float()
        self.D4_2 = nn.Linear(input_size, hidden_size, bias=False)
        self.D4_2.weight.data = torch.abs(self.D4_2.weight.data).float()
        ##
        ##
        self.D5_2 = nn.Linear(hidden_size, hidden_size)
        self.D5_2.weight.data = torch.abs(self.D5_2.weight.data).float()
        self.D6_2 = nn.Linear(input_size, hidden_size, bias=False)
        self.D6_2.weight.data = torch.abs(self.D6_2.weight.data).float()
        ##
        ##
                ## _3
        self.W_ii_3 = nn.Linear(input_size, hidden_size, bias=False)
        self.W_ii_3.weight.data = torch.abs(self.W_ii_3.weight.data).float()
        self.W_hi_3 = nn.Linear(hidden_size, hidden_size, bias=False)
        self.W_hi_3.weight.data = torch.abs(self.W_hi_3.weight.data).float()
        self.b_i_3 = nn.Parameter(torch.zeros(1, hidden_size))

        self.W_if_3 = nn.Linear(input_size, hidden_size, bias=False)
        self.W_if_3.weight.data = torch.abs(self.W_if_3.weight.data).float()
        self.W_hf_3 = nn.Linear(hidden_size, hidden_size, bias=False)
        self.W_hf_3.weight.data = torch.abs(self.W_hf_3.weight.data).float()
        self.b_f_3 = nn.Parameter(torch.zeros(1, hidden_size))

        self.W_ig_3 = nn.Linear(input_size, hidden_size, bias=False)
        self.W_ig_3.weight.data = torch.abs(self.W_ig_3.weight.data).float()
        self.W_hg_3 = nn.Linear(hidden_size, hidden_size, bias=False)
        self.W_hg_3.weight.data = torch.abs(self.W_hg_3.weight.data).float()
        self.b_g_3 = nn.Parameter(torch.zeros(1, hidden_size))

        self.W_io_3 = nn.Linear(input_size, hidden_size, bias=False)
        self.W_io_3.weight.data = torch.abs(self.W_io_3.weight.data).float()
        self.W_ho_3 = nn.Linear(hidden_size, hidden_size, bias=False)
        self.W_ho_3.weight.data = torch.abs(self.W_ho_3.weight.data).float()
        self.b_o_3 = nn.Parameter(torch.zeros(1, hidden_size))

        self.D1_3 = nn.Linear(input_size, hidden_size, bias=False)
        self.D1_3.weight.data = torch.abs(self.D1_3.weight.data).float()
        self.D2_3 = nn.Linear(input_size, hidden_size, bias=False)
        self.D2_3.weight.data = torch.abs(self.D2_3.weight.data).float()
        self.D3_3 = nn.Linear(input_size, hidden_size, bias=False)
        self.D3_3.weight.data = torch.abs(self.D3_3.weight.data).float()
        self.D4_3 = nn.Linear(input_size, hidden_size, bias=False)
        self.D4_3.weight.data = torch.abs(self.D4_3.weight.data).float()
        ##
        ##
        self.D5_3 = nn.Linear(hidden_size, hidden_size)
        self.D5_3.weight.data = torch.abs(self.D5_3.weight.data).float()
        self.D6_3 = nn.Linear(input_size, hidden_size, bias=False)
        self.D6_3.weight.data = torch.abs(self.D6_3.weight.data).float()
        ##
        ##
                ## _4
        self.W_ii_4 = nn.Linear(input_size, hidden_size, bias=False)
        self.W_ii_4.weight.data = torch.abs(self.W_ii_4.weight.data).float()
        self.W_hi_4 = nn.Linear(hidden_size, hidden_size, bias=False)
        self.W_hi_4.weight.data = torch.abs(self.W_hi_4.weight.data).float()
        self.b_i_4 = nn.Parameter(torch.zeros(1, hidden_size))

        self.W_if_4 = nn.Linear(input_size, hidden_size, bias=False)
        self.W_if_4.weight.data = torch.abs(self.W_if_4.weight.data).float()
        self.W_hf_4 = nn.Linear(hidden_size, hidden_size, bias=False)
        self.W_hf_4.weight.data = torch.abs(self.W_hf_4.weight.data).float()
        self.b_f_4 = nn.Parameter(torch.zeros(1, hidden_size))

        self.W_ig_4 = nn.Linear(input_size, hidden_size, bias=False)
        self.W_ig_4.weight.data = torch.abs(self.W_ig_4.weight.data).float()
        self.W_hg_4 = nn.Linear(hidden_size, hidden_size, bias=False)
        self.W_hg_4.weight.data = torch.abs(self.W_hg_4.weight.data).float()
        self.b_g_4 = nn.Parameter(torch.zeros(1, hidden_size))

        self.W_io_4 = nn.Linear(input_size, hidden_size, bias=False)
        self.W_io_4.weight.data = torch.abs(self.W_io_4.weight.data).float()
        self.W_ho_4 = nn.Linear(hidden_size, hidden_size, bias=False)
        self.W_ho_4.weight.data = torch.abs(self.W_ho_4.weight.data).float()
        self.b_o_4 = nn.Parameter(torch.zeros(1, hidden_size))

        self.D1_4 = nn.Linear(input_size, hidden_size, bias=False)
        self.D1_4.weight.data = torch.abs(self.D1_4.weight.data).float()
        self.D2_4 = nn.Linear(input_size, hidden_size, bias=False)
        self.D2_4.weight.data = torch.abs(self.D2_4.weight.data).float()
        self.D3_4 = nn.Linear(input_size, hidden_size, bias=False)
        self.D3_4.weight.data = torch.abs(self.D3_4.weight.data).float()
        self.D4_4 = nn.Linear(input_size, hidden_size, bias=False)
        self.D4_4.weight.data = torch.abs(self.D4_4.weight.data).float()
        ##
        ##
        self.D5_4 = nn.Linear(hidden_size, hidden_size)
        self.D5_4.weight.data = torch.abs(self.D5_4.weight.data).float()
        self.D6_4 = nn.Linear(input_size, hidden_size, bias=False)
        self.D6_4.weight.data = torch.abs(self.D6_4.weight.data).float()
        ##
        ##
                ## _5
        self.W_ii_5 = nn.Linear(input_size, hidden_size, bias=False)
        self.W_ii_5.weight.data = torch.abs(self.W_ii_5.weight.data).float()
        self.W_hi_5 = nn.Linear(hidden_size, hidden_size, bias=False)
        self.W_hi_5.weight.data = torch.abs(self.W_hi_5.weight.data).float()
        self.b_i_5 = nn.Parameter(torch.zeros(1, hidden_size))

        self.W_if_5 = nn.Linear(input_size, hidden_size, bias=False)
        self.W_if_5.weight.data = torch.abs(self.W_if_5.weight.data).float()
        self.W_hf_5 = nn.Linear(hidden_size, hidden_size, bias=False)
        self.W_hf_5.weight.data = torch.abs(self.W_hf_5.weight.data).float()
        self.b_f_5 = nn.Parameter(torch.zeros(1, hidden_size))

        self.W_ig_5 = nn.Linear(input_size, hidden_size, bias=False)
        self.W_ig_5.weight.data = torch.abs(self.W_ig_5.weight.data).float()
        self.W_hg_5 = nn.Linear(hidden_size, hidden_size, bias=False)
        self.W_hg_5.weight.data = torch.abs(self.W_hg_5.weight.data).float()
        self.b_g_5 = nn.Parameter(torch.zeros(1, hidden_size))

        self.W_io_5 = nn.Linear(input_size, hidden_size, bias=False)
        self.W_io_5.weight.data = torch.abs(self.W_io_5.weight.data).float()
        self.W_ho_5 = nn.Linear(hidden_size, hidden_size, bias=False)
        self.W_ho_5.weight.data = torch.abs(self.W_ho_5.weight.data).float()
        self.b_o_5 = nn.Parameter(torch.zeros(1, hidden_size))

        self.D1_5 = nn.Linear(input_size, hidden_size, bias=False)
        self.D1_5.weight.data = torch.abs(self.D1_5.weight.data).float()
        self.D2_5 = nn.Linear(input_size, hidden_size, bias=False)
        self.D2_5.weight.data = torch.abs(self.D2_5.weight.data).float()
        self.D3_5 = nn.Linear(input_size, hidden_size, bias=False)
        self.D3_5.weight.data = torch.abs(self.D3_5.weight.data).float()
        self.D4_5 = nn.Linear(input_size, hidden_size, bias=False)
        self.D4_5.weight.data = torch.abs(self.D4_5.weight.data).float()
        ##
        ##
        self.D5_5 = nn.Linear(hidden_size, hidden_size)
        self.D5_5.weight.data = torch.abs(self.D5_5.weight.data).float()
        self.D6_5 = nn.Linear(input_size, hidden_size, bias=False)
        self.D6_5.weight.data = torch.abs(self.D6_5.weight.data).float()
        ##


        self.W_yh = nn.Linear(hidden_size, output_size)
        self.b_y = nn.Parameter(torch.zeros(1, output_size))

        self.g1 = nn.Tanh()
        self.g2 = ShiftedReLU(beta)
        ##
    def save_weights(self, file_path):

        torch.save(self.state_dict(), file_path)
        self.save_path = file_path
        print("Model weights saved successfully.")
    def load_weights(self, file_path):
        try:
            self.load_state_dict(torch.load(file_path))
            self.save_path = file_path
            print("Model weights loaded successfully.")
        except Exception as e:
            print("Error while loading model weights:", str(e))
    def forward(self, input, hidden, cell_state, input_t_past):
        i_t_1 = self.g1(self.W_ii_1(input) + self.W_hi_1(hidden) + self.D1_1(input_t_past) + self.b_i_1)
        i_t_1 = torch.clamp(i_t_1, min=0, max=None)
        f_t_1 = self.g1(self.W_if_1(input) + self.W_hf_1(hidden) + self.D2_1(input_t_past) + self.b_f_1)
        f_t_1 = torch.clamp(f_t_1, min=0, max=None)
        g_t_1 = self.g1(self.W_ig_1(input) + self.W_hg_1(hidden) + self.D3_1(input_t_past) + self.b_g_1)
        g_t_1 = torch.clamp(g_t_1, min=0, max=None)
        o_t_1 = self.g1(self.W_io_1(input) + self.W_ho_1(hidden) + self.D4_1(input_t_past) + self.b_o_1)
        o_t_1 = torch.clamp(o_t_1, min=0, max=None)
        cell_state_1 = f_t_1 * cell_state + i_t_1 * g_t_1
        h_t_1 = o_t_1 * self.g2(cell_state_1+ self.D5_1(hidden) + self.D6_1(input))
        ##
        ##
                ##_2
        i_t_2 = self.g1(self.W_ii_2(input) + self.W_hi_2(h_t_1) + self.D1_2(input_t_past) + self.b_i_2)
        i_t_2 = torch.clamp(i_t_2, min=0, max=None)
        f_t_2 = self.g1(self.W_if_2(input) + self.W_hf_2(h_t_1) + self.D2_2(input_t_past) + self.b_f_2)
        f_t_2 = torch.clamp(f_t_2, min=0, max=None)
        g_t_2 = self.g1(self.W_ig_2(input) + self.W_hg_2(h_t_1) + self.D3_2(input_t_past) + self.b_g_2)
        g_t_2 = torch.clamp(g_t_2, min=0, max=None)
        o_t_2 = self.g1(self.W_io_2(input) + self.W_ho_2(h_t_1) + self.D4_2(input_t_past) + self.b_o_2)
        o_t_2 = torch.clamp(o_t_2, min=0, max=None)
        cell_state_2 = f_t_2 * cell_state_1 + i_t_2 * g_t_2
        h_t_2 = o_t_2 * self.g2(cell_state_2+ self.D5_2(h_t_1) + self.D6_2(input))
                ##
                ##_3
        i_t_3 = self.g1(self.W_ii_3(input) + self.W_hi_3(h_t_2) + self.D1_3(input_t_past) + self.b_i_3)
        i_t_3 = torch.clamp(i_t_3, min=0, max=None)
        f_t_3 = self.g1(self.W_if_3(input) + self.W_hf_3(h_t_2) + self.D2_3(input_t_past) + self.b_f_3)
        f_t_3 = torch.clamp(f_t_3, min=0, max=None)
        g_t_3 = self.g1(self.W_ig_3(input) + self.W_hg_3(h_t_2) + self.D3_3(input_t_past) + self.b_g_3)
        g_t_3 = torch.clamp(g_t_3, min=0, max=None)
        o_t_3 = self.g1(self.W_io_3(input) + self.W_ho_3(h_t_2) + self.D4_3(input_t_past) + self.b_o_3)
        o_t_3 = torch.clamp(o_t_3, min=0, max=None)
        cell_state_3 = f_t_3 * cell_state_2 + i_t_3 * g_t_3
        h_t_3 = o_t_3 * self.g2(cell_state_3+ self.D5_3(h_t_2) + self.D6_3(input))
        ##
                ##_4
        i_t_4 = self.g1(self.W_ii_4(input) + self.W_hi_4(h_t_3) + self.D1_4(input_t_past) + self.b_i_4)
        i_t_4 = torch.clamp(i_t_4, min=0, max=None)
        f_t_4 = self.g1(self.W_if_4(input) + self.W_hf_4(h_t_3) + self.D2_4(input_t_past) + self.b_f_4)
        f_t_4 = torch.clamp(f_t_4, min=0, max=None)
        g_t_4 = self.g1(self.W_ig_4(input) + self.W_hg_4(h_t_3) + self.D3_4(input_t_past) + self.b_g_4)
        g_t_4 = torch.clamp(g_t_4, min=0, max=None)
        o_t_4 = self.g1(self.W_io_4(input) + self.W_ho_4(h_t_3) + self.D4_4(input_t_past) + self.b_o_4)
        o_t_4 = torch.clamp(o_t_4, min=0, max=None)
        cell_state_4 = f_t_4 * cell_state_3 + i_t_4 * g_t_4
        h_t_4 = o_t_4 * self.g2(cell_state_4+ self.D5_4(h_t_3) + self.D6_4(input))
        ##
                       ## 5
                        ## _5
        i_t_5 = self.g1(self.W_ii_5(input) + self.W_hi_5(h_t_4) + self.D1_5(input_t_past) + self.b_i_5)
        i_t_5 = torch.clamp(i_t_5, min=0, max=None)
        f_t_5 = self.g1(self.W_if_5(input) + self.W_hf_5(h_t_4) + self.D2_5(input_t_past) + self.b_f_5)
        f_t_5 = torch.clamp(f_t_5, min=0, max=None)
        g_t_5 = self.g1(self.W_ig_5(input) + self.W_hg_5(h_t_4) + self.D3_5(input_t_past) + self.b_g_5)
        g_t_5 = torch.clamp(g_t_5, min=0, max=None)
        o_t_5 = self.g1(self.W_io_5(input) + self.W_ho_5(h_t_4) + self.D4_5(input_t_past) + self.b_o_5)
        o_t_5 = torch.clamp(o_t_5, min=0, max=None)
        cell_state_5 = f_t_5 * cell_state_4 + i_t_5 * g_t_5
        h_t_5 = o_t_5 * self.g2(cell_state_5+ self.D5_5(h_t_4) + self.D6_5(input))
        ##
        ##
        ##
        output = self.W_yh(h_t_5) + self.b_y
        return output, h_t_5, cell_state_5



def train_loop(dataloader, model, loss_fn, optimizer, writer, epoch):
    model.train()
    total_loss = 0.0

    for input, target in dataloader:
        optimizer.zero_grad()

        batch_size = input.size(0)
        sequence_length = input.size(1)
        hidden_state = torch.zeros(batch_size, model.hidden_size)
        cell_state = torch.zeros(batch_size, model.hidden_size)  # New addition for LSTM
        input_t_past = torch.zeros_like(input[:, 0])  # Initialize input_t_past with zeros

        for t in range(sequence_length):
            output, hidden_state, cell_state = model(input[:, t], hidden_state, cell_state, input_t_past)
            loss = loss_fn(output, target[:, t])
            total_loss += loss.item()
            input_t_past = input[:, t]  # Update input_t_past for the next time step
        loss.backward()
        optimizer.step()
    writer.add_scalar('Loss/train', total_loss / len(dataloader), epoch)
    writer.flush()
    return total_loss / len(dataloader)


def test_loop(dataloader, model, loss_fn, writer, epoch):
    model.eval()
    total_loss = 0.0
    with torch.no_grad():
        for input, target in dataloader:
            batch_size = input.size(0)
            sequence_length = input.size(1)
            hidden_state = torch.zeros(batch_size, model.hidden_size)
            cell_state = torch.zeros(batch_size, model.hidden_size)  # New addition for LSTM
            input_t_past = torch.zeros_like(input[:, 0])  # Initialize input_t_past with zeros

            for t in range(sequence_length):
                output, hidden_state, cell_state = model(input[:, t], hidden_state, cell_state, input_t_past)
                loss = loss_fn(output, target[:, t])
                total_loss += loss.item()
                input_t_past = input[:, t]  # Update input_t_past for the next time step
    writer.add_scalar('Loss/test', total_loss / len(dataloader), epoch)
    writer.flush()
    return total_loss / len(dataloader)

def call_train_loop(epochs, train_loader, test_loader, model, loss_fn, optimizer, writer):
    train_losses = []
    test_losses = []
    for epoch in range(epochs):
        train_loss = train_loop(train_loader, model, loss_fn, optimizer, writer, epoch)
        test_loss = test_loop(test_loader, model, loss_fn, writer, epoch)
        train_losses.append(train_loss)
        test_losses.append(test_loss)
        print(f"Epoch {epoch+1}/{epochs} - Train Loss: {train_loss:.4f}, Test Loss: {test_loss:.4f}")
    for epoch in range(epochs):
        writer.add_scalars('Loss', {'Train': train_losses[epoch], 'Test': test_losses[epoch]}, epoch + 1)
    writer.close()
    return train_losses, test_losses



#  LSTM count_trainable_params

def count_trainable_params(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)




def train_LSTM(learning_rate, epochs, train_input, train_target, test_input, test_target, model_save_path1, model_save_path2):
    allComperatinRequired=[]
    input_size = 5
    hidden_size = 72
    sequence_length = 9
    output_size = 1
    batch_size = 24
    time_string = time.strftime("%Y%m%d-%H%M%S")
    writer = SummaryWriter()

    # Prepare data
    train_input_expanded = train_input.reshape(-1, sequence_length, input_size)
    train_target_expanded = train_target.reshape(-1, sequence_length, output_size)

    model = LSTM(input_size, hidden_size, output_size)  # Creating LSTM model
    # Example usage
    #lstm_model = LSTM(input_size=5, hidden_size=72, output_size=1)  # Replace with your actual sizes
    Number_trainable_parameters=count_trainable_params(model)
    print("Number of trainable parameters:", Number_trainable_parameters)
    allComperatinRequired.append(Number_trainable_parameters)
    print(allComperatinRequired)
    ##
    ##
    ##
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    train_size = int(0.8 * len(train_input_expanded))
    test_size = len(train_input_expanded) - train_size
    train_dataset = torch.utils.data.TensorDataset(train_input_expanded[:train_size], train_target_expanded[:train_size])
    test_dataset = torch.utils.data.TensorDataset(train_input_expanded[train_size:], train_target_expanded[train_size:])
    train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=False)
    test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

    train_losses, test_losses = call_train_loop(epochs, train_loader, test_loader, model, criterion, optimizer, writer)
    writer.close()
    train_losses_min = np.min(train_losses)
    test_losses_min = np.min(test_losses)
    allComperatinRequired.append(train_losses_min)
    allComperatinRequired.append(test_losses_min)
    print(allComperatinRequired)


    prediction_seqs, avg_cvrmse = evaluate_model_2(model, test_input, test_target, sequence_length, input_size)
    allComperatinRequired.append(avg_cvrmse)
    print(allComperatinRequired)
    prediction_seqs, avg_MAPE = evaluate_model_mape(model, test_input, test_target, sequence_length, input_size)
    allComperatinRequired.append(avg_MAPE)
    print(allComperatinRequired)

    # Save the model and plot the training and test losses
    torch.save(model.state_dict(), model_save_path1)
    torch.save(model.state_dict(), model_save_path2)

    image_name = "LSTM_losses.jpg"
    new_image_name = f"{time_string}_{image_name}"
    plt.plot(range(1, epochs + 1), train_losses, label='Train Loss')
    plt.plot(range(1, epochs + 1), test_losses, label='Test Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.grid()
    plt.legend()
    plt.savefig(new_image_name)
    plt.show()
    return allComperatinRequired

class ShiftedReLU(nn.Module):
    def __init__(self, beta):
        super(ShiftedReLU, self).__init__()
        self.relu = nn.ReLU()
        self.beta = beta

    def forward(self, x):
        return self.relu(x) - self.beta


if __name__ == "__main__":
    beta = 100
    model_namE = "ICLSTM"
    new_model_namE = f"{time_string}_{model_namE}"
    model_save_path1 = new_model_namE + ".pt"
    model_save_path2 = new_model_namE + ".h5"

    # Convert train_input and train_target to float32 type
    train_input = train_input.float()
    train_target = train_target.float()
    test_input = test_input.float()
    test_target = test_target.float()

    # Call the train_RNN function and specify the model save path
    allComperatinRequired_Eng=train_LSTM(1e-3, 100, train_input, train_target, test_input, test_target, model_save_path1, model_save_path2)
```

```python
def evaluate_model(model, test_input, test_target, sequence_length, input_size):
    model.eval()
    with torch.no_grad():
        prediction_seqs = []
        accuracy_sum = 0.0
        for i in range(len(test_input)):
            input_seq = test_input[i:i+1]
            target_seq = test_target[i:i+1]

            input_expanded = input_seq.reshape(1, sequence_length, input_size)
            hidden_state = torch.zeros(1, model.hidden_size)
            cell_state = torch.zeros(1, model.hidden_size)  # Fix batch size to 1 for LSTM
            input_t_past = torch.zeros_like(input_expanded[:, 0])
            prediction_seq = []

            for t in range(sequence_length):
                output, _, hidden_state = model(input_expanded[:, t], hidden_state, cell_state, input_t_past)
                prediction_seq.append(output[0, 0])  # Append the model output directly

                input_t_past = input_expanded[:, t]

            prediction_seq = np.array(prediction_seq)
            prediction_seqs.append(prediction_seq)  # Append the whole sequence

            target_seq = target_seq.squeeze().numpy()

            accuracy = 1.0 - np.abs(prediction_seq - target_seq).mean() / np.abs(target_seq).mean()
            accuracy_sum += accuracy

            plt.plot(range(sequence_length), target_seq, label='Target')
            plt.plot(range(sequence_length), prediction_seq, label='Prediction')
            plt.xlabel('Time Step')
            plt.ylabel('Energy Usage')
            plt.legend()
            plt.grid()
            plt.show()

    return prediction_seqs
```

```python
def evaluate_model(model, test_input, test_target, sequence_length, input_size, Verbose):
    model.eval()
    with torch.no_grad():
        prediction_seqs = []
        accuracy_sum = 0.0
        for i in range(len(test_input)):
            input_seq = test_input[i:i+1]
            target_seq = test_target[i:i+1]

            input_expanded = input_seq.reshape(1, sequence_length, input_size)
            hidden_state = torch.zeros(1, model.hidden_size)
            cell_state = torch.zeros(1, model.hidden_size)  # Fix batch size to 1 for LSTM
            input_t_past = torch.zeros_like(input_expanded[:, 0])
            prediction_seq = []

            for t in range(sequence_length):
                output, _, hidden_state = model(input_expanded[:, t], hidden_state, cell_state, input_t_past)
                prediction_seq.append(output[0, 0])  # Append the model output directly

                input_t_past = input_expanded[:, t]

            prediction_seq = np.array(prediction_seq)
            prediction_seqs.append(prediction_seq)  # Append the whole sequence

            target_seq = target_seq.squeeze().numpy()

            accuracy = 1.0 - np.abs(prediction_seq - target_seq).mean() / np.abs(target_seq).mean()
            accuracy_sum += accuracy
            if (Verbose):
                plt.plot(range(sequence_length), target_seq, label='Target')
                plt.plot(range(sequence_length), prediction_seq, label='Prediction')
                plt.xlabel('Time Step')
                plt.ylabel('Energy Usage')
                plt.legend()
                plt.grid()
                plt.show()

        average_accuracy = accuracy_sum / len(test_input)
        if (Verbose):print("Average Accuracy on Test Data:", average_accuracy)
        return np.array(prediction_seqs)

```

```python
def scaled_gradient(x, predictions, target, tset, tref_high, tref_low):
    mis_tracking_target=tf.square(predictions- target)
    print(" \n mis_tracking_target :",  mis_tracking_target )
    grad, = tf.gradients(mis_tracking_target, x)
    grad_comfort_high = 1/((tref_high - tset))
    grad_comfort_low = 1/((tset - tref_low))
    grad_contrained = grad[:,0:2,0:1] + 0.000005*(grad_comfort_high+grad_comfort_low)
    print(" \n")
    return grad, grad_contrained
```
